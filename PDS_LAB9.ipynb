```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/wine_quality.csv")

sns.pairplot(df, vars=["fixed_acidity", "volatile_acidity", "citric_acid", "residual_sugar", "alcohol"], hue="quality")
plt.suptitle("Wine Quality Assessment - Pair Plot", y=1.02)
plt.show()
```
