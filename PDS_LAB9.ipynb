```python
# Lab 9 - Scenario 6: Wine Quality Assessment
# Author: Group Members (23CS201SE4 02)
# Course: Python for Data Science

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create dataset
data = {
    'fixed acidity': [7.4, 7.8, 7.8, 11.2, 7.4, 7.9, 6.7, 7.3, 7.8, 6.6],
    'volatile acidity': [0.70, 0.88, 0.76, 0.28, 0.70, 0.60, 0.58, 0.65, 0.58, 0.60],
    'citric acid': [0.00, 0.00, 0.04, 0.56, 0.00, 0.06, 0.08, 0.00, 0.02, 0.10],
    'residual sugar': [1.9, 2.6, 2.3, 1.9, 1.9, 1.6, 1.8, 1.2, 2.0, 1.5],
    'density': [0.9978, 0.9968, 0.9970, 0.9980, 0.9962, 0.9964, 0.9959, 0.9956, 0.9976, 0.9958],
    'quality': [5, 5, 5, 6, 5, 6, 6, 5, 5, 6]
}

df = pd.DataFrame(data)

# Display dataset
print("Wine Quality Dataset:")
print(df.head())

# Create pair plot
sns.pairplot(df, hue='quality', palette='coolwarm', diag_kind='kde')
plt.suptitle("Pair Plot: Wine Quality Analysis", y=1.02)
plt.show()
```
