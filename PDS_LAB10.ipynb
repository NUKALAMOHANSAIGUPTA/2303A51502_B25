```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/exam_results.csv")

sns.set_theme(style="darkgrid")  # Theme
sns.catplot(x="class", y="exam_score", hue="gender", kind="box", data=df, palette="Set2")
plt.title("Exam Results by Class and Gender")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/housing_data.csv")

corr = df[["area","bhk","bath","price","age"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Housing Feature Correlation Matrix")
plt.show()
```
