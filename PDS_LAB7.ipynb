```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/co2_emissions.csv")

plt.plot(df['Year'], df['CO2_Emissions'], marker='o', linestyle='-', color='b')
plt.title("Global CO2 Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (in million metric tons)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/student_scores.csv")

plt.plot(df['Exam_Number'], df['Score'], marker='o', linestyle='-', color='g')
plt.title("Student Performance Over Exams")
plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```
